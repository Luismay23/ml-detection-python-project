```python
# To label the datasets
```

```python
benign_files = sorted([p for p in device_folder.glob("*.csv")])
```

```python
gafgyt_folder = device_folder / "Gafgyt"
gafgyt_file = gafgyt_folder / "combo.csv"
```

```python
mirai_folder = device_folder / "Mirai"
mirai_files = ["ack.csv", "scan.csv", "syn.csv", "udp.csv", "udpplain.csv"]
mirai_files = [mirai_folder / f for f in mirai_files if (mirai_folder / f).exists()]
```

```python
dfs = []
```

```python
#This will be for labeling
```

```python
#For the benign files
```

```python
for f in benign_files:
    df = pd.read_csv(f)
    df['label'] = 'benign'
    dfs.append(df)
```

```python
#For the mirai files
```

```python
for f in mirai_files:
    df = pd.read_csv(f)
    df['label'] = 'mirai'
    dfs.append(df)
```

```python
#For the gafgyt files
```

```python
df = pd.read_csv(gafgyt_file)
df['label'] = 'gafgyt'
dfs.append(df)
```

```python
# This is the process for combining it into 1 single dataset that the ML models can read
```

```python
combined = pd.concat(dfs, ignore_index=True)
print("Combined dataframe shape:", combined.shape)
print("Class distribution:\n", combined['label'].value_counts())
```

```python
combined.shape
```

```python
combined.head()
```

```python
# I've decided to truncate this to only 100k rows because of the file size
```

```python
MAX_ROWS = 100_000
small_combined = combined.sample(n=MAX_ROWS, random_state=42)
```

```python
zip_path = "/content/combined_danmini.zip"
csv_filename_in_zip = "combined_danmini.csv"  # the filename inside the zip
```

```python
small_combined.to_csv(zip_path, index=False, compression={'method':'zip', 'archive_name': csv_filename_in_zip})
print(f" Saved truncated & compressed dataset to {zip_path}")
```

```python
file_path = Path("/content/combined_danmini.zip")
size_mb = file_path.stat().st_size / (1024*1024)
print(f"Size of file: {size_mb:.2f} MB")
```
