```python
#The Random Forest baseline ML Implementation :
```

```python
import time
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    precision_recall_curve,
    auc
)
```

```python
#dataFrame to dataFrame 3 will be for the Danmini doorbell
```

```python
dataFrame = pd.read_csv("/content/Danmini/combined_danmini.csv")
```

```python
print("Dataset shape:", dataFrame.shape)
print(dataFrame.head())
```

```python
X = dataFrame.drop(columns=["label"])
y = dataFrame["label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
#I decided to scale to make it more fair for the large numeric ranges
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#The training of Random Forest
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)
```

```python
#The time for it more specifically
```

```python
start_train = time.time()
rf.fit(X_train_scaled, y_train)
end_train = time.time()


print(f"\nTraining time: {end_train - start_train:.3f} seconds")

```

```python
#Interference latency measuring :
```

```python
start_pred = time.time()
y_pred = rf.predict(X_test_scaled)
end_pred = time.time()

latency = (end_pred - start_pred) / len(X_test)
print(f"Inference latency: {latency*1000:.6f} ms per sample")
```

```python
#Classification
```

```python
#f1 score is here
```

```python
print(classification_report(y_test, y_pred))
```

```python
#Confusion Matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
#ROC-AUC
```

```python
y_proba = rf.predict_proba(X_test_scaled)

roc_auc = roc_auc_score(
    y_test,
    y_proba,
    multi_class="ovr",
    average="macro"
)

print(f"\nMacro ROC-AUC: {roc_auc:.4f}")
```

```python
#PR-AUC
```

```python
labels = sorted(y.unique())
y_test_bin = pd.get_dummies(y_test)[labels]
pr_aucs = []

for i, cls in enumerate(labels):
    precision, recall, _ = precision_recall_curve(y_test_bin.iloc[:, i], y_proba[:, i])
    pr_auc = auc(recall, precision)
    pr_aucs.append(pr_auc)

macro_pr_auc = np.mean(pr_aucs)
print(f"Macro PR-AUC: {macro_pr_auc:.4f}")
```

```python
#Proposed model : Dense Neural Network (MLP) [This is the deep learning model] for Danmini
```

```python
dataFrame3 = pd.read_csv("/content/Danmini/combined_danmini.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
#I reused the previous importing lines for ROC - AUC, PR-AUC, and f1 score
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X = dataFrame3.drop(columns="label", axis=1)
y = dataFrame3["label"]
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
num_classes = len(np.unique(y_encoded))
y_onehot = np.eye(num_classes)[y_encoded]
```

```python
#Scaling
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#MLP Split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_onehot, test_size=0.2, random_state=42, stratify=y_onehot
)
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(num_classes, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#MLP training
```

```python
#I made it so the max is 50 rounds (epochs)
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
start_train = time.time()
history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50,
    batch_size=128,
    callbacks=[early_stop],
    verbose=2
)
train_time = time.time() - start_train
```

```python
start_infer = time.time()
y_pred_probs = model.predict(X_test)
inference_time = time.time() - start_infer
y_pred_classes = np.argmax(y_pred_probs, axis=1)
y_test_classes = np.argmax(y_test, axis=1)
```

```python
#ROC - AUC
```

```python
roc_auc = roc_auc_score(y_test, y_pred_probs, multi_class="ovr")
```

```python
#PR - AUC
```

```python
pr_auc = average_precision_score(y_test, y_pred_probs)
```

```python
#F1
```

```python
f1 = f1_score(y_test_classes, y_pred_classes, average='weighted')
```

```python
#Latency measuring
```

```python
latency_per_sample_ms = (inference_time / X_test.shape[0]) * 1000
```

```python
print("MLP Dense Neural Network for Danmini")
print(f"ROC-AUC:        {roc_auc:.4f}")
print(f"PR-AUC:         {pr_auc:.4f}")
print(f"F1 Score:       {f1:.4f}")
print(f"Train Time:     {train_time:.4f} seconds")
print(f"Inference Time (total): {inference_time:.4f} seconds")
print(f"Inference Latency (per sample): {latency_per_sample_ms:.6f} ms")
```
